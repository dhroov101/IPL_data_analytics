```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import csv
```

```python
source=requests.get('https://www.espncricinfo.com/records/trophy/team-match-results/indian-premier-league-117')
```

```python
soup=bs(source.content,'html.parser')
```

```python
table=soup.find('tbody')
print(table)
```

```python
for row in table.findAll('tr'):
    print(row.prettify())
    break
```

```python
import re
matches=[]
for row in table.findAll('tr'):
    a_list=[]
    for td in row:
        a_list.append(td)
    match_x=str((a_list[-1]))
    result = re.search('-(.*)full-scorecard"', match_x)
    match=result.group(1)
    match=match[-8:-1]
    if match[0]=='-':
        match=match[1:8]
    matches.append(match)
    
```

```python
print(matches)
```

```python
match=soup.findAll('tr')
print(match)
```

```python
filename='IPL2023_match_results.csv'
with open(filename, 'w', newline='') as f:
    for match.tr in match:
        for td in match.tr:
            f.write(td.text+',')
        f.write('\n')
f.close()
```
